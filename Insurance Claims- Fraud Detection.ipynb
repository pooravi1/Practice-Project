Data Processing

```python
#Conveting fraud_reported's data type from Object to int
df.replace({'fraud_reported':{'Y':1,'N':0}}, inplace=True)
df.head()
```

```python
df[['insured_zip']] = df[['insured_zip']].astype(object)
```

```python
df = df.drop(columns=['policy_number', 'insured_zip', 'policy_bind_date', 'incident_date', 'incident_location','auto_year',
                      'incident_hour_of_the_day','policy_state','insured_sex','insured_education_level','insured_occupation'])
df.head()
```

we will drop Collision Type as well.

```python
df.drop('collision_type',axis=1,inplace=True)
df.head()
```

Applying one-hot encoding to convert all categorical variables except out target variables

```python
dummies = pd.get_dummies(df[[
    'policy_csl',  
    'insured_hobbies', 
    'insured_relationship',
    'incident_type', 
    'incident_severity',
    'authorities_contacted', 
    'incident_state', 
    'incident_city',
    'auto_make', 
    'auto_model']])

dummies = dummies.join(df[[
    'property_damage', 
    'police_report_available', 
    "fraud_reported"]])

dummies.head()
```

```python
df.dtypes
```

```python
df = pd.concat([df, df._get_numeric_data()], axis=1)  # joining numeric columns
df.head()
```

```python
df = df.drop(columns=['policy_csl','insured_hobbies','insured_relationship','incident_type','incident_severity',
                      'authorities_contacted','incident_state','incident_city'])
df.head()
```

```python
#Splitting the independent and target variables in x and y before removing the skewness
x=df.drop("fraud_reported",axis=1)
y=df["fraud_reported"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
#K Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
knn.score(x_train,y_train)
predknn=knn.predict(x_test)
print(accuracy_score(y_test,predknn))
print(confusion_matrix(y_test,predknn))
print(classification_report(y_test,predknn))
```

```python
#Random Forest Classfier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```
