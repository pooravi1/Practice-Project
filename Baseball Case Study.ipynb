Splitting Data into feature and target variable.

```python
bb.columns
```

```python
x=bb.drop("W",axis=1)
y=bb["W"]
```

```python
x.shape
```

```python
y.shape
```

Linear Reqression

```python
#Importing Required Libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
#Splitting the data into train and validation
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=50)
```

```python
#Building a model
lr = LinearRegression(normalize=True)
```

```python
#train the model
lr.fit(x_train,y_train)
```

```python
predictions = lr.predict(x_test)
```

```python
mae = mean_absolute_error(y_test,predictions)
mae
```

```python
lr.score(x_train,y_train)
```

Ridge Regression 

```python
#importing library
from sklearn.linear_model import RidgeCV
```

```python
rm = RidgeCV(alphas=(0.01,0.1,1,10),normalize=True)
```

```python
#training the model
rm.fit(x_train,y_train)
```

```python
predictions = rm.predict(x_test)
```

```python
mae_rm = mean_absolute_error(y_test,predictions)
mae_rm
```

```python
rm.score(x_train,y_train)
```

Lasso Regression

```python
#importing library
from sklearn.linear_model import Lasso
```

```python
ls = Lasso(alpha =(0.01),normalize=True)
```

```python
#training the model
ls.fit(x_train,y_train)
```

```python
predictions = ls.predict(x_test)
```

```python
mae_ls = mean_absolute_error(y_test,predictions)
mae_ls
```

```python
ls.score(x_train,y_train)
```

As we can see Linear Regression has good score. now we will check the cross-validation score and select the best model.

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(lr,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(rm,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(ls,x,y,cv=5)
print(score)
print(score.mean())
```

By comparing all cross validation score we can see we received the max score in Linear Regaression. So that Linear Regression is the best model to predict the Wins.

LinearRegression does not have hyperparameters that can be tuned. So we will not perform Hyper parameter tuning.
